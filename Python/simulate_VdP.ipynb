# Simulation of the Van der Pol oscillator

... using parameters/settings from "Sparse reconstruction of ordinary differentail equations with inference", Sara Venkatraman et al.

... using instructions from https://www.johndcook.com/blog/2019/12/22/van-der-pol/

```python
from scipy import linspace
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def vdp(t, z):
    x, y = z
    return [y, mu*(1 - x**2)*y - x]
```

```python
a, b = 0, 15
mu = 2
h = 0.1
t = linspace(a, b, int(b/h))
sol = solve_ivp(vdp, [a, b], [1, 0], t_eval=t)
```

```python
plt.axes().set_aspect(1)
plt.plot(sol.t, sol.y[0], label = "x1")
plt.plot(sol.t, sol.y[1], label = "x2")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

### Save file

```python
# save time series as csv file
filepath = "../data/"
df_ts = pd.DataFrame({'t': sol.t, 'x1': sol.y[0], 'x2': sol.y[1]})
print(df_ts[0:10])
df_ts.to_csv(f'{filepath}ts_VanderPol_python.csv', index=False)
```
